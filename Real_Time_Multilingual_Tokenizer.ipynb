```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, pipeline
from deepmultilingualpunctuation import PunctuationModel
from typing import List
from punctuators.models import SBDModelONNX
import gradio as gr
import nltk
```

```python
model = PunctuationModel(model="oliverguhr/fullstop-punctuation-multilang-large")
```

```python
model1 = PunctuationModel(model="kredor/punctuate-all")
```

```python
sm = SBDModelONNX.from_pretrained("sbd_multi_lang")
```

```python
iii = 'I am living in istanbul where are you'
```

```python
mm =  model1.restore_punctuation(iii)
mm
```

```python
def tokenizer(text):
  p_text = model.restore_punctuation(text)
  input_texts: List[str] = [p_text]
  results_pun: List[List[str]] = sm.infer(input_texts)
  doc = ''
  for input_text, output_texts in zip(input_texts, results_pun):
    for text in output_texts:
      doc += text

  return doc

def real_time_type(text):
    text = tokenizer(text)
    
    return text

gr.Interface(
    fn=real_time_type, 
    inputs=gr.Textbox( max_lines=20,streaming=True), 
    outputs="textbox", 
    live=True).launch()
```
